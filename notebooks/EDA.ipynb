```python
import numpy as np
import pandas as pd 
import scipy as sp 
import matplotlib.pyplot as plt
import pylab as pl
import plotly_express as px
pd.options.plotting.backend= 'hvplot'
```

```python
data_car_sales=pd.read_csv('vehicles_us.csv')
```

```python
data_car_sales.info()
```

```python
data_car_sales['model_year'].sample()
```

Printing info and converting each category to proper Dtypes in order to properly compare.

```python
data_car_sales.isna().sum()
```

printing isna to make sure there is no missing information in categories that will matter when comparing data

```python
data_car_sales.dropna(subset=['model_year'],inplace=True)
#data_car_sales['model_year']=data_car_sales['model_year'].round()
data_car_sales['model_year'].isna().sum()
```

```python
data_car_sales['model_year']=data_car_sales['model_year'].astype('int')
```

```python
data_car_sales.dropna(subset=['cylinders'],inplace=True)
data_car_sales['cylinders'].isna().sum()
data_car_sales['cylinders']=data_car_sales["cylinders"].astype('int')
```

```python
data_car_sales.dropna(subset=['odometer'],inplace=True)
data_car_sales['odometer'].isna().sum()
data_car_sales['odometer']=data_car_sales['odometer'].astype('int')
```

```python
data_car_sales.dropna(subset=['paint_color'],inplace=True)
data_car_sales.dropna(subset=['is_4wd'],inplace=True)
data_car_sales['is_4wd'].replace({1.0: 'yes'})
data_car_sales['is_4wd'].astype('string')
```

```python
data_car_sales.sample()
```

```python
#data_car_sales['date_posted']=data_car_sales['date_posted'].astype('int')
data_car_sales['date_posted']=pd.to_datetime(data_car_sales['date_posted'])
```

```python
data_car_sales.info()
data_car_sales.isna().sum()
```

```python
data_car_sales['manufacturer']=data_car_sales['model'].apply(lambda x: x.split()[0])
data_car_sales
```

Above you will see the data is cleaned up containing the correct Dtype format and zero for null values.
